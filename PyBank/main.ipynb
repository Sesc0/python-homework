```python
# Importing necessary libraries
from pathlib import Path
import numpy as np
import csv
```

```python
# Setting the path for the budget data csv
csvpath = Path(r'02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv')
```

```python
with open(csvpath,'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=',')
    csvheader = next(csvreader)
    months = []
    pnl = []
    for row in csvreader:
        months.append(row[0])
        pnl.append(int(row[1]))
```

```python
change = []
for index in range(len(pnl)-1):
    change.append(pnl[index+1]-pnl[index])
max_index = np.argmax(change)
min_index = np.argmin(change)
```

```python
string1 = f'Total Months: {len(months)}\n'
string2 = f'Total: ${sum(pnl)}\n'
string3 = f'Average change: ${round(sum(change)/len(change),2)}\n'
string4 = f'Greatest increase in profits: ${months[max_index+1]} ${change[max_index]}\n'
string5 = f'Greatest decrease in profits: ${months[min_index+1]} $({change[min_index]})\n'
strings = [string1, string2, string3, string4, string5]
```

```python
output_file = open('output.txt','w')
output_file.write('Financial Analysis\n')
output_file.write('-'*20 + '\n')
output_file.writelines(strings)
output_file.close()
```
