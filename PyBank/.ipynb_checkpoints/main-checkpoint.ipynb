```python
# Importing necessary libraries
from pathlib import Path
import csv
```

```python
# Setting the path for the budget data csv
csvpath = Path(r'02-Homework_02-Python_Instructions_PyBank_Resources_budget_data.csv')
```

```python
with open(csvpath,'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=',')
    csvheader = next(csvreader)
    months = []
    pnl = []
    for row in csvreader:
        months.append(row[0])
        pnl.append(row[1])
```

```python
with open(csvpath,'r') as csvfile:
    csvreader = csv.reader(csvfile,delimiter=',')
    csvheader = next(csvreader)
    budget_dict = dict((rows[0],[int(rows[1])]) for rows in csvreader)
budget_dict
```

```python
first_line = True
for key,value in budget_dict.items():
    if first_line == True:
        difference = 0
        old_value = value[0]
        first_line =  False
    else:
        difference = old_value - value[0]
        old_value = value[0]
    budget_dict[key].append(difference)
budget_dict
```

```python
for key in budget_dict:
    print(key.value())
```
